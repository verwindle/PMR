# To find T2 of H2O 

```python
ln_t2_h2o = log(t2_h2o['field'])
plot(ln_t2_h2o, 'k')
xlabel(', ', fontsize=20)
ylabel(' , %', fontsize=20)
suptitle('Natural logarithm of H2O 90 relaxation ', fontsize=30)
```

# Linear extrapolation

# We'll take only straight part, not exactly linearising

```python
ln_t2_h2o[5000:6000]
```

```python
ln_t2_h2o = log(t2_h2o['field'][:5330])  # guess 5330's a good end point
x = t2_h2o[:5330].index
y = ln_t2_h2o
lin = interp1d(x, y, kind='linear')  
ln_t2_h2o_straight = lin(t2_h2o[:5330].index)
plot(x, y, 'o', x, ln_t2_h2o_straight, '-k')
k, b = polyfit(x, y, 1)[0], polyfit(x, y, 1)[1]  # koeff in y = kx+b
xlabel(', ', fontsize=20)
ylabel(' , %', fontsize=20)
suptitle('Natural logarithm of H2O 90 relaxation linear part', fontsize=30)
```

```python
x0 = t2_h2o.index
y0 = k*x0 + b
x = t2_h2o.index
y = ln_t2_h2o
plot(x0, y0, 'b', x, y, '-r')
xlabel(', ', fontsize=20)
ylabel(' , %', fontsize=20)
suptitle('Natural logarithm of H2O 90 relaxation linearized', fontsize=30)
patch = mpt.Patch(color='red', label='-t/T2')
patch_l = mpt.Patch(color='blue', label='linearized')
patch_res = mpt.Patch(color='white', label='T2 = 2474 c')
legend(handles=[patch, patch_l, patch_res], fontsize=30)
```

```python
t2 = -1/k
```

```python
t2 = Symbol('T2')
display(Eq(t2, 2474))
```
