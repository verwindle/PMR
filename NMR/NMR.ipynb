# Comparison plot

```python
fids = {'H2O': h2o['field'],
    'H2O_5': h2o_5['field'],
    'Mn_20mkl': mn_20['field'],
    'Mn_40mkl': mn_40['field'],
    'Mn_80mkl': mn_80['field'],
    'Na_20mkl' : na_20['field'],
    'Na_40mkl' : na_40['field']}
```

```python
nms = list(fids.keys())
labels = list(fids.values())
colors = ['red', 'blue', 'green', 'magenta', 'yellow', 'cyan', 'black']
plot(h2o.index, labels[0], '-r', labels[1], '-b', labels[2], '-g', labels[3], '-m', \
     labels[4], '-y', labels[5], '-c', labels[6], '-k')
#fuck
r_patch = mpt.Patch(color=colors[0], label=nms[0])
b_patch = mpt.Patch(color=colors[1], label=nms[1])
g_patch = mpt.Patch(color=colors[2], label=nms[2])
m_patch = mpt.Patch(color=colors[3], label=nms[3])
y_patch = mpt.Patch(color=colors[4], label=nms[4])
c_patch = mpt.Patch(color=colors[5], label=nms[5])
k_patch = mpt.Patch(color=colors[6], label=nms[6])
legend(handles=[r_patch, b_patch, g_patch, m_patch, y_patch, c_patch, k_patch], fontsize=20)
suptitle('Fids comparison', fontsize=30)
```
